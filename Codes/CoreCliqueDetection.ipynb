```python
import networkx as nx
import matplotlib.pyplot as plt

# Recreate the graph based on the uploaded image
G = nx.Graph()

# Define the edges based on the network
edges = [
    ('A', 'B'), ('A', 'C'), ('A', 'D'),('A', 'E'), ('A', 'F'),
    ('F', 'A'), ('J', 'F'), ('F', 'K'), ('F', 'L'), ('F', 'G'),
    ('K', 'J'), ('K', 'N'), ('K', 'M'), ('K', 'G'),
    ('G', 'L'), ('G', 'H'), ('H', 'L'), ('I', 'H'), ('L', 'M'),
    ('L', 'N'), ('N', 'M'), ('M', 'P'), ('P', 'O'), ('O', 'N')

]
G.add_edges_from(edges)

# Compute the largest 3-core sub-network
k_core_subnetwork = nx.k_core(G, k=3)

# Plot the largest 3-core sub-network
plt.figure(figsize=(4, 4))
nx.draw(k_core_subnetwork, with_labels=True, node_color='green', node_size=700, font_size=12, font_weight='bold', edge_color='black')
plt.title("Largest 3-Core Sub-Network")
plt.show()
```

```python
# Recreate the graph based on the provided image
G_clique = nx.Graph()
n_clique = 3

# Define the edges from the uploaded network
edges_clique = [
    ('A', 'B'), ('A', 'C'), ('A', 'D'),('A', 'E'), ('A', 'F'),
    ('F', 'A'), ('J', 'F'), ('F', 'K'), ('F', 'L'), ('F', 'G'),
    ('K', 'J'), ('K', 'N'), ('K', 'M'), ('K', 'G'),
    ('G', 'L'), ('G', 'H'), ('H', 'L'), ('I', 'H'), ('L', 'M'),
    ('L', 'N'), ('N', 'M'), ('M', 'P'), ('P', 'O'), ('O', 'N')

]

G_clique.add_edges_from(edges_clique)

# Finding the largest 3-clique sub-network
cliques = list(nx.find_cliques(G_clique))
three_cliques = [clique for clique in cliques if len(clique) == n_clique]

# Plot the largest 3-clique sub-network
plt.figure(figsize=(4, 4))
largest_3_clique = max(three_cliques, key=len)  # Get the largest n-clique
subgraph_3_clique = G_clique.subgraph(largest_3_clique)

# Plot the subgraph corresponding to the 3-clique
nx.draw(subgraph_3_clique, with_labels=True, node_color='green', node_size=700, font_size=12, font_weight='bold', edge_color='black')
plt.title("Largest 3-Clique Sub-Network")
plt.show()

# Return the nodes and edges of the largest 3-clique sub-network
subgraph_3_clique.nodes(), subgraph_3_clique.edges()
```
